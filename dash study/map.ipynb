```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
def first_map():
    df = pd.read_csv(r'C:\Users\kofij\Desktop\UI Data 2.csv')
    df['text'] = df['Area_name'] + '<br>Civilian_labor_force_2019: ' + df['Civilian_labor_force_2019'] + ', Unemployment_Rate: ' + (df['Unemployment_rate_2019']).astype(str)+ '% ' + '<br>Civilian_labor_force_2020: ' + df['Civilian_labor_force_2020'] + ', Unemployment_Rate: ' +(df['Unemployment_rate_2020']).astype(str)+ '% ' + '<br>Median_Household_Income_2019: ' + df['Median_Household_Income_2019']
    limits = [(50,1000),(50,1000),(200,2000),(200,2000),(1000,5000)] # for the size of the bubbles
    colors = ["blue","black","white","orange","yellow"] # colors of the bubbles
    cities = []
    scale = 80

    fig = go.Figure()

    df['Civilian_labor_force_2019'] = pd.to_numeric(df.Civilian_labor_force_2019, errors='coerce').fillna(0).astype(np.int64)

    for i in range(len(limits)):
        lim = limits[i]
        df_sub = df[lim[0]:lim[1]]
        fig.add_trace(go.Scattergeo(
            locationmode = 'USA-states',
            lon = df_sub['lon'],
            lat = df_sub['lat'],
            text = df_sub['text'],
            marker = dict(
                size = df_sub['Civilian_labor_force_2019']/scale,
                color = colors[i],
                line_color='rgb(40,40,40)',
                line_width=0.5,
                sizemode = 'area'
            ),
            name = '{0} - {1}'.format(lim[0],lim[1])))

    fig.update_layout(
            title_text = 'state unemployment data before pandemic<br>from 2000 to 2019',
            showlegend = True,
            geo = dict(
                scope = 'usa',
                landcolor = 'rgb(118, 186, 193)',
            )
        )
    return fig.show()
```
